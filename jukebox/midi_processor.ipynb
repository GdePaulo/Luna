```python
pm = pmid.PrettyMIDI('happy-birthday.mid')
mainInstrument = 0
pitchShift = 0

city = pmid.PrettyMIDI('city-of-stars.mid')
mainInstrument = 1
pitchShift = 4

love = pmid.PrettyMIDI('love-of-my-life.mid')
mainInstrument = 0
pitchShift = 0

pm = love
# Let's look at what's in this MIDI file
print('There are {} time signature changes'.format(len(pm.time_signature_changes)))
print('There are {} instruments'.format(len(pm.instruments)))
print('Instrument 3 has {} notes'.format(len(pm.instruments[0].notes)))
print('Instrument 4 has {} pitch bends'.format(len(pm.instruments[0].pitch_bends)))
print('Instrument 5 has {} control changes'.format(len(pm.instruments[0].control_changes)))
```

```python
for instrument in pm.instruments:
    print(instrument)

result = ""

lastStart = 0
lastEnd = 0
# for i in range(0, len(pm.instruments[3].notes)): 
for i in range(0, len(pm.instruments[mainInstrument].notes)): 

    note = pm.instruments[mainInstrument].notes[i]
    note.pitch += pitchShift
    print(f"pitch: {note.pitch} | start: {note.start} | end: {note.end}")

    if (note.start == lastStart):
        print("Skipping....")
        # print(f"before {result}")
        result = ",".join(result.split(",")[:-2]) 
        if (result):
            result += ","
        # print(f"after {result}")
        # continue
    elif(note.start > lastEnd):
        diff = int((note.start - lastEnd) * 1000)
        result += f"0-{diff},"

    lastStart = note.start
    lastEnd = note.end

    freq = pmid.note_number_to_hz(note.pitch)
    dur = int((note.end - note.start) * 1000)
    result += f"{int(freq)}-{dur}"
    if i < len(pm.instruments[mainInstrument].notes) - 1:
        result += ","
    # print(f"current:{result}")
    # print(pmid.note_number_to_name(note.pitch))
print(result)

import clipboard as c
c.copy(result)
```
